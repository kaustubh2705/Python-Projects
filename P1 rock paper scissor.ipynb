```python
import random
computer_wins = 0
player_wins = 0
winning = int(input("Enter winning number"))
print("Type 'quit' or 'q' or exit the game")
while player_wins < winning and computer_wins < winning:
    print(f"player score {player_wins} and computer score {computer_wins}")
    print("...rock")
    print("...paper")
    print("...scissor")

    player = input('Enter your choice:').lower()
    if player == "quit" or player =="q":
        break
    rand_num = random.randint(0,2)
    if rand_num == 0:
        computer = 'rock'
    elif rand_num == 1:
        computer = 'paper'
    else:
        computer = 'scissor'

    #print('computer plays {}'.format(computer))  # for python 3.6.4
    print(f"computer plays {computer}")  # for python version 3.6.5

    if player:
        if player == computer:
            print("It's a tie")
        elif player == "rock":
            if computer == "scissor":
                print("You wins")
                player_wins += 1
            elif computer == "paper":
                print("Computer wins!!")
                computer_wins += 1
        elif player =="paper":
            if computer == "rock":
                print("You wins")
                player_wins += 1
            elif computer == "scissor":
                print("Computer wins!!")
                computer_wins += 1
        elif player == "scissor":
            if computer == "paper":
                print("You wins")
                player_wins += 1
            elif computer == "rock":
                print("Computer wins!!")
                computer_wins += 1
        else:
            print("OOPS Something went wrong!!")
            
print(f"Final score:player score {player_wins} and computer score {computer_wins}")
if player_wins > computer_wins:
    print("You won the match!!")
elif player_wins < computer_wins:
    print("You Lose :( !!")
else:
    print("It's a Tie!!")
print("Thanks for Playing... develop by Kaustubh Khandelwal")
```
